**Imbalanced Dataset:**: A datasset with an unequal class distribution

```python
import pandas as pd
import numpy as np
```

```python
#loading the dataset into the pandas dataframe
credit_data=pd.read_csv('/content/credit_data.csv')
```

```python
credit_data.head()
```

```python
credit_data.shape
```

```python
credit_data.isnull().sum()
```

```python
#distribution of the two classses
credit_data['Class'].value_counts()
```

this is highly imbalanced dataset

*   0 -> Legit Transaction
*   1 -> Fraudulent Transaction



```python
#separeting the legit and Fraudulent
legit=credit_data[credit_data.Class==0]
fraud=credit_data[credit_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

under-sampling

Bulid the sample dataset containing similar distribution of Legit & Fradulent Transactionds

```python
legit_sample=legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the two DataFrames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```
