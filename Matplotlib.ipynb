About Matplotlib
Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations. It provides a comprehensive range of tools for plotting various types of graphs and charts, making it a go-to library for data visualization in Python.
Commonly Used Modules:
matplotlib.pyplot: Provides a MATLAB-like interface for creating plots.
matplotlib.animation: Used to create animations.
matplotlib.style: Enables the use of predefined styles for plots.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#sin wave
plt.plot(x,y)
plt.show()
```

```python
#cos wave
plt.plot(x,z)
plt.show()
```

```python
#adding x-axis & y-axis labels
plt.title("Sin Wave")
plt.xlabel("angles")
plt.ylabel("sin values")
plt.plot(x,y)
plt.show()
```

```python
#adding title & x-axis & y-axis to cos wave
plt.title("Cos Wave")
plt.xlabel("X-Axis")
plt.ylabel("Cos values")
plt.plot(x,z)
plt.show()
```

```python
#Parabila
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
plt.plot(x,y,'b*')
plt.show()
```

```python
#plotting two in one graph
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'b*')
plt.plot(x,np.cos(x),'g-')
plt.show()
```

Bar Plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
la=['Telugu','Hindi','English','Marati']
people=[95,98,99,28]
ax.bar(la,people)
plt.xlabel('la')
plt.ylabel('people')
plt.show()
```

Pie Charts

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
la=['Telugu','hindi','English','Marti']
people=[99,150,100,28]
ax.pie(people,labels=la,autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x=np.linspace(0,10,20)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter plots

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
x=20*np.random.random(100)
y=np.sin(x)
z=np.cos(x)
ax.scatter(x,y,z,c=z,cmap='Greens')
plt.show()
```
