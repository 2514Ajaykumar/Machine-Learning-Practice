```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data collection and processing

```python
car_dataset=pd.read_csv('/content/car data.csv')
```

```python
car_dataset.head()
```

```python
car_dataset.shape
```

```python
# getting some information about the dataset
car_dataset.info()
```

```python
car_dataset.isnull().sum()
```

```python
#checking the distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Encodering the catagorical data

```python
# encodering "Fuel_type" column
car_dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

# encodering "Seller" column
car_dataset.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

# encodering "Transmission" column
car_dataset.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
car_dataset.head()
```

Train and test data

```python
X=car_dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y=car_dataset['Selling_Price']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

Linear Regression

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

Model evaluation

```python
#r2 Error
#Prediction on training data
training_data_prediction=model.predict(X_train)
train_accuracy=metrics.r2_score(Y_train,training_data_prediction)
```

```python
print("Accuracy on training data:",train_accuracy)
```

Visualize the acutal price and prdicted price

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
```

```python
#r2 Error
#Prediction on test data
test_data_prediction=model.predict(X_test)
test_accuracy=metrics.r2_score(Y_test,test_data_prediction)
```

```python
print("Accuracy on test data:",test_accuracy)
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
```

Lasso Regression

```python
Lasso_model=Lasso()
```

```python
Lasso_model.fit(X_train,Y_train)
```

```python
#prediction on training data
training_data_prediction=Lasso_model.predict(X_train)
train_accuracy=metrics.r2_score(Y_train,training_data_prediction)
```

```python
print("Accuracy on training data:",train_accuracy)
```

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
```

```python
#prediction on testing data
test_data_prediction=Lasso_model.predict(X_test)
test_accuracy=metrics.r2_score(Y_test,test_data_prediction)
```

```python
print("Accuracy on test data:",test_accuracy)
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
```
